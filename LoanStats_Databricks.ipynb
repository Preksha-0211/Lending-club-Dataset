## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/tables/LoanStats.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Create a view or table

temp_table_name = "LoanStats"

df.createOrReplaceTempView(temp_table_name)
```

```
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.feature import OneHotEncoder
CategoricalCols = ['grade','home_ownership','purpose']
```

```
# To handle the categorical variables
# applied for loop to go through all the columns and applied OneHotEncoder, StringIndexer
stages = []
for CateCols in CategoricalCols:
    stringIndexer = StringIndexer(inputCol = CateCols, outputCol = CateCols + "Index")
    encoder = OneHotEncoder(inputCols = [stringIndexer.getOutputCol()], outputCols = [CateCols + "catVec"])
    stages += [stringIndexer, encoder]
```

```
stages
```

### Deisgn Matrix
Function 'VectorAssembler'

```
from pyspark.ml.feature import VectorAssembler
numericCols = ['loan_amnt','term','annual_inc']
assembleInputs = assemblerInputs = [c + "catVec" for c in CategoricalCols] + numericCols
VectorAssemblerData = VectorAssembler(inputCols = assembleInputs, outputCol = "features")
stages += [VectorAssemblerData]
```

```
label_Idx = StringIndexer(inputCol = 'int_rate', outputCol = 'label')
stages += [label_Idx]
```

```
# Creating a pipeline, when feeding the data operate all the steps like vectore, onehotencoder....
pipeline = Pipeline().setStages(stages)
pipelineModel = pipeline.fit(df)
```

```
# Apply to data
# transforming the datase (train and text)
InputMLModeldf = pipelineModel.transform(df)
```

```
InputMLModeldf.toPandas()
```

```
## Spliting the datset
from pyspark.sql.functions import split, col
DF_train, DF_test = InputMLModeldf.randomSplit(weights = [0.7,0.3], seed = 123)
```

### Machine learning model

```
# Importing linear regression model and training it on test data
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol = 'features', labelCol='label', maxIter=10, regParam=0.3, elasticNetParam=0.8)
lr_model = lr.fit(DF_train)
print("Coefficients: " + str(lr_model.coefficients))
print("Intercept: " + str(lr_model.intercept))
```

###Summarize the model over the training set and print out some metrics:

```
trainingSummary = lr_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)     


#RMSE measures the differences between predicted values by the model and the actual values. 
#However, RMSE alone is meaningless until we compare with the actual int_rate value, such as mean, min and max. After such comparison, our RMSE looks pretty good.
```

```
DF_train.describe().show()
```

```
lr_predictions = lr_model.transform(DF_test)
lr_predictions.select("prediction","int_rate","features").show(5)
from pyspark.ml.evaluation import RegressionEvaluator
lr_evaluator = RegressionEvaluator(predictionCol="int_rate", \
                 labelCol="features",metricName="r2")
print("R Squared (R2) on test data = %g" % lr_evaluator.evaluate(lr_predictions))
```

```
test_result = lr_model.evaluate(DF_test)
print("Root Mean Squared Error (RMSE) on test data = %g" % test_result.rootMeanSquaredError)
print("r2: %f" % test_result.r2) 
```

## Random Forest regression

```
from pyspark.ml.regression import RandomForestRegressor
dt = RandomForestRegressor(featuresCol ='features', labelCol = 'label')
dt_model = dt.fit(DF_train)
dt_predictions = dt_model.transform(DF_test)

dt_evaluator = RegressionEvaluator(
    labelCol="int_rate", predictionCol="prediction", metricName="rmse")
rmse = dt_evaluator.evaluate(dt_predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)

dt_evaluator_r2 = RegressionEvaluator(
    labelCol="features", predictionCol="int_rate", metricName="r2")
r2 = dt_evaluator_r2.evaluate(dt_predictions)
print("r2: %f" % r2) 
```

###### RMSE has increased

```
from pyspark.ml.tuning import ParamGridBuilder
import numpy as np

paramGrid = ParamGridBuilder() \
    .addGrid(dt.numTrees, [int(x) for x in np.linspace(start = 10, stop = 50, num = 3)]) \
    .addGrid(dt.maxDepth, [int(x) for x in np.linspace(start = 5, stop = 25, num = 3)]) \
    .build()
```

```
from pyspark.ml.tuning import CrossValidator
from pyspark.ml.evaluation import RegressionEvaluator
crossval = CrossValidator(estimator=dt,
                          estimatorParamMaps=paramGrid,
                          evaluator=RegressionEvaluator(),
                          numFolds=3)
```

```
cvModel = crossval.fit(DF_train)
predictions = cvModel.transform(DF_test)
```

```
# eVALUATE
import matplotlib.pyplot as plt
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
rfPred = cvModel.transform(InputMLModeldf)
rfResult = rfPred.toPandas()
plt.plot(rfResult.label, rfResult.prediction, 'bo')
plt.xlabel('int_rate')
plt.ylabel('Prediction')
plt.suptitle("Model Performance RMSE: %f" % rmse)
plt.show()
```
